```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import requests
from math import sqrt
from bs4 import BeautifulSoup
from IPython.display import Markdown, HTML

%matplotlib notebook
```

```python
data1 = pd.read_csv('result_df.csv')
cor1 = data1.corr()
def partial_cor(a,b,c):
    rdk = ((a-b*c)/ 
       (sqrt(1-b**2)*sqrt(1-c**2)))
    return rdk

def pcor(comt,covariate):
    rd={}
    mylist=list(comt.columns.values)
    if covariate in mylist:
        mylist.remove(covariate)
    #list of all column names, remove covariate from that list,
    #pass to that corrl vation
    for i in mylist:
        rd[i]={}
        for j in mylist:
            if i==j:
                rd[i][j]=1
            else:
                rd[i][j]=partial_cor(comt.loc[i,j],comt.loc[i,covariate],comt.loc[j,covariate])
    rdf=pd.DataFrame(rd)
    return rdf
pcor(cor1,'CAI')
```
